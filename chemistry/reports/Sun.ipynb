## 3. The function defined in Python 3
In this section, functions are defined to accelerate this process, which will

(1) Retrieve data from OOI data server using API  
(2) Extract vocabulary metadata  
(3) Extract deployment information  
(4) Extract calibration information  
(5) Extract annotation  
(6) Generate figures  

Please note that it will take a while to pull NetCDF files and process all data.

```python
# Set up Python 3 environment
import requests
import datetime
import matplotlib.pyplot as plt
import re
import time
import os
import datetime
import matplotlib.dates as dates
import numpy as np
import xarray as xr
import seaborn as sns
import pandas as pd
sns.set()
```

```python
# Enter API authentification information
# Use your own username and token
username =''
token= ''
```

```python
# Default base URL. DO NOT CHANGE.
DATA_API = 'https://ooinet.oceanobservatories.org/api/m2m/12576/sensor/inv'
VOCAB_API = 'https://ooinet.oceanobservatories.org/api/m2m/12586/vocab/inv'
ASSET_API = 'https://ooinet.oceanobservatories.org/api/m2m/12587'
```

```python
# Generate URL template. DO NOT CHANGE.
url_template = DATA_API+\
    '/%s/'+\
    '%s/'+\
    '%s/'+\
    'recovered_host/'+\
    'ctdmo_ghqr_sio_mule_instrument'
```

```python
# Specify functions to convert timestamps
ntp_epoch = datetime.datetime(1900, 1, 1)
unix_epoch = datetime.datetime(1970, 1, 1)
ntp_delta = (unix_epoch - ntp_epoch).total_seconds()

def ntp_seconds_to_datetime(ntp_seconds):
    return datetime.datetime.utcfromtimestamp(ntp_seconds - ntp_delta).replace(microsecond=0)
  
def convert_time(ms):
  if ms != None:
    return datetime.datetime.utcfromtimestamp(ms/1000)
  else:
    return None
```

```python
# Define an integrated function
def get_data(instruments, measurement, unit, beginDT, endDT):
    
    # Create an empty figure
    line_color = ['b','r']
    plt.close()
    fig, (upper,lower) = plt.subplots(2,1,figsize=(16,10), sharex=True, sharey=False)

    # Request data from OOI data server
    for j in range(len(instruments)):
        
# Vocabulary Metadata
        data_request_url_v ='/'.join((VOCAB_API,instruments[j][0],instruments[j][1],instruments[j][2]))
        # print(data_request_url_v) # debug only
        r = requests.get(data_request_url_v, auth=(username, token))
        if j == 0:
            data_v_0 = r.json()
        else:
            data_v_1 = r.json()

# Deployment Information
        # Setup the API request url
        data_request_url_d = ASSET_API + '/events/deployment/query'
        params = {
            'beginDT':beginDT,
            'endDT':endDT,
            'refdes':instruments[j][0]+'-'+instruments[j][1]+'-'+instruments[j][2],   
        }

        # Grab the information from the server
        r = requests.get(data_request_url_d, params=params, auth=(username, token))
        data = r.json()

        df = pd.DataFrame() # Setup empty array
        for d in data:
            df = df.append({
                'deployment': d['deploymentNumber'],
                'start': convert_time(d['eventStartTime']),
                'stop': convert_time(d['eventStopTime']),
                'latitude': d['location']['latitude'],
                'longitude': d['location']['longitude'],
                'sensor': d['sensor']['uid'],
                'asset_id': d['sensor']['assetId'],
            }, ignore_index=True)
        if j == 0:
            data_d_0 = df
        else:
            data_d_1 = df

# Calibration Information
        # Setup the API request url
        data_request_url_c = ASSET_API + '/asset/cal'
        params = {
            'beginDT':beginDT,
            'endDT':endDT,
            'refdes':instruments[j][0]+'-'+instruments[j][1]+'-'+instruments[j][2],
        }

        # Grab the information from the server
        r = requests.get(data_request_url_c, params=params, auth=(username, token))
        data = r.json()

        # Reformat the data into a pretty table
        df = pd.DataFrame() # Setup empty array
        for d in data:
            for dd in d['sensor']['calibration']:
                for ddd in dd['calData']:
                    df = df.append({
                        'value': ddd['value'],
                        'start': convert_time(ddd['eventStartTime']),
                        'stop': convert_time(ddd['eventStopTime']),
                        'name': ddd['eventName'],
                        'assetUid': ddd['assetUid'],
                    }, ignore_index=True)
    
        df = df.sort_values(by=['start','name'])
        if j == 0:
            data_c_0 = df
        else:
            data_c_1 = df

# Annotation
        ANNO_API = 'https://ooinet.oceanobservatories.org/api/m2m/12580/anno/find'
        # Note: CHANGE DATES HERE MANUALLY
        params = {
            'beginDT':int(datetime.date(2013,1,1).strftime('%s'))*1000,
            'endDT':int(datetime.date(2017,12,31).strftime('%s'))*1000,
            'refdes':instruments[j][0]+'-'+instruments[j][1]+'-'+instruments[j][2],
        }

        r = requests.get(ANNO_API, params=params, auth=(username, token))
        data = r.json()

        df = pd.DataFrame() # Setup empty array
        for d in data:
            df = df.append({
                'annotation': d['annotation'],
                'start': convert_time(d['beginDT']),
                'stop': convert_time(d['endDT']),
                'site': d['subsite'],
                'node': d['node'],
                'sensor': d['sensor'],
                'id': d['id']
            }, ignore_index=True)
        pd.set_option('display.max_colwidth', -1) # Show the full annotation text
        if j == 0:
            data_a_0 = df
        else:
            data_a_1 = df

# Generate Plots
        # Extract the URL
        params = {
            'beginDT':beginDT,
            'endDT':endDT,
        }
        data_request_url = url_template % (instruments[j][0],instruments[j][1],instruments[j][2])
        # print(data_request_url) # debug only
        r = requests.get(data_request_url, params=params, auth=(username, token))
        data = r.json()
        print(instruments[j])
        # print(len(data)) # debug only
        url = data['allURLs'][0]
        # print(url) # debug only
    
        # Set time delay for OOI server to prepare all files. Unit: second
        time.sleep(90)
    
        # Extract all .nc files from the link and remove all other useless files
        tds_url = 'https://opendap.oceanobservatories.org/thredds/dodsC'
        datasets = requests.get(url).text
        urls = re.findall(r'href=[\'"]?([^\'" >]+)', datasets)
        x = re.findall(r'(ooi/.*?.nc)', datasets)
        for i in x:
            if i.endswith('.nc') == False:
                x.remove(i)
        for i in x:
            try:
                float(i[-4])
            except:
                x.remove(i)
        datasets = [os.path.join(tds_url, i) for i in x]
        # print(datasets) # debug only
    
        # Aggregate data
        ds = xr.open_mfdataset(datasets)
        ds = ds.swap_dims({'obs': 'time'})
        ds = ds.chunk({'time': 100})
        if j == 0:
            ds_0 = ds
        else:
            ds_1 = ds
    
        # Reshape data
        for i in [0, 1]:
            %%time
            from dask.diagnostics import ProgressBar
            with ProgressBar():
                df = ds[measurement[i]].to_dataframe()
                df_daily = df.resample('D').mean() # daily average
    
            # Plot
            %%time
            # df[measurement].plot(ax=figure) # raw data
            if i == 0:
                df_daily[measurement[i]].plot(ax=upper, color=line_color[j]) # daily average
            else:
                df_daily[measurement[i]].plot(ax=lower, color=line_color[j]) # daily average
    
        # Modify the appearance of the figure
        upper.set_ylabel(unit[0])
        lower.set_ylabel(unit[1])
        lower.set_xlabel('Time')
        upper.legend(['Flanking Subsurface Mooring A', 'Flanking Subsurface Mooring B'])
        lower.legend(['Flanking Subsurface Mooring A', 'Flanking Subsurface Mooring B'])
        plt.suptitle('Global Station Papa @ ' + str(instruments[0][3]) + ' m');
    
    # Save figure
    plt.savefig('Global_PP_' + str(instruments[0][3]) + '_m.jpeg')
    
    # Do not show figures right now
    plt.close()
    
    # Save all data as corresponding variables
    return data_v_0, data_d_0, data_c_0, data_a_0, ds_0, fig, data_v_1, data_d_1, data_c_1, data_a_1, ds_1
```

## 6. Comparison of CTD time series from Flanking Mooring A & B 

This section compares the temperature and practical salinity data between Mooring A and Mooring B at each depth. All plots are based on the daily averaged value.

### 6.1 Depth: 30 m

```python
fig_3
```

Since mooring A and mooring B are close to each other, the temperature and salinity at both sites at the same time point should be quite similar. As is shown in the figure above, most data from mooring A and B matches very well and the trend is correct, especially the salinity data. However, some abnormal data is observed and key findings are as follows.

(1) Three abnormal straight line in salinity figure. These lines appear at the beginning of three deployments because CTD was turned on before lowered into the water, which recorded data from the surface to the designated depth. Other issues, such as air bubbles and particles may exist in the CTD pump upon deployment.  

(2) As is indicated in the annotation, the anti-biofouling shield was missed upon recovery of the platform and biofouling was evident on shallow CTDs for both CTDs. However, the influence of biofouling cannot be assessed due to the lack of recovered data of Deployment #2.  

(3) The depths of Deployment 1~4 are 5~15 meters deeper than expected. The influence of changes of depths cannot be evaluated on the figure since CTDs on both moorings are deployed deeper at the same time. The temperature difference between A and B may be caused by currents since the distance between two sites is ~40 km.

## 8. Notable results & actions needed

In conclusion, the overall CTD data quality at Flanking Mooring A & B at Global Station Papa is satisfying. By conducting the comparison, data from Mooring A and B matches and trends are correct. However, some aspects still need to be paid attention to.  

(1) The data during the first day of each deployment should not be used. This is because CTD was turned on upon lowered into the water, which recorded all data from the surface to deeper depth. Thus, this influence on data is less obvious on shallow depths but is significantly evident on deep depths.  

(2) The annotation is detailed. However, point (1) above needs to be added so that other researchers can be aware of this issue.

(3) Further calibration with shipboard CTD data is needed, which is the most efficient way to evaluate the accuracy of these datasets.  

(4) Recovered data of Deployment #2 is missing.

(5) The influence of biofouling needs to be evaluated when the missing data is available.

Additionally, the function defined above may help other scientists to extract data quickly. Please feel free to share the code with colleagues.
